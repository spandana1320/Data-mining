```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("/content/sample_stock_data.csv")

# Select numerical features for PCA
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA_5', 'MA_10']
x = df[features]

# Standardize the features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x_scaled)

# Create a DataFrame with principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Company'] = df['Company']

# Display the PCA result
print(pca_df.head())

# Optional: Visualize the PCA result
plt.figure(figsize=(8, 6))
for company in pca_df['Company'].unique():
    subset = pca_df[pca_df['Company'] == company]
    plt.scatter(subset['PC1'], subset['PC2'], label=company)
plt.title('PCA of Stock Market Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("sample_stock_data.csv")

# Map companies to industry sectors
sector_map = {
    'AlphaCorp': 'Technology',
    'BetaTech': 'Healthcare'
}
df['Sector'] = df['Company'].map(sector_map)

# Select numerical features for PCA
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA_5', 'MA_10']
x = df[features]

# Standardize the features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x_scaled)

# Create a DataFrame with principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Sector'] = df['Sector']

# Plot the PCA result by sector
plt.figure(figsize=(8, 6))
for sector in pca_df['Sector'].unique():
    subset = pca_df[pca_df['Sector'] == sector]
    plt.scatter(subset['PC1'], subset['PC2'], label=sector)
plt.title('PCA of Stock Market Data by Sector')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
